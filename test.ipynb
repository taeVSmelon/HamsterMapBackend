```python
from pymongo import MongoClient
import time

client = MongoClient('mongodb+srv://admin:mark6506@cluster0.0tcdjdu.mongodb.net/PHC_TestV2')
```

```python
db = client['PHC_TestV2']
collection = db['UserData']

# อัปเดต field ที่กำหนดไว้เท่านั้น
results = collection.find()
cheatUsers = []

for result in results:
    username = result["username"]
    oldScore = result["score"]
    print(username, oldScore)
    clearedStages = result["stats"]["clearedStages"]
    myClearedStages = {}
    subtractScores = {}
    
    for game, stages in clearedStages.items():
        if isinstance(stages, list):
            myClearedStages[game] = []
            for stage in stages:
                haveStageId = len([s for s in myClearedStages[game] if s["stageId"] == stage["stageId"]]) > 0
                if not haveStageId:
                    myClearedStages[game].append(stage)
                else:
                    if username not in cheatUsers:
                        cheatUsers.append(username)
                    if game in subtractScores:
                        subtractScores[game] += 10
                    else:
                        subtractScores[game] = 10
        else:
            myClearedStages[game] = stages
    
    for game, score in subtractScores.items():
        # print(username, game, score)
        collection.update_one(
            {"username": username},
            {
                "$inc": {f"score.{game}": -score},
                "$set": {f"stats.clearedStages.{game}": myClearedStages[game]}
            }
        )

print(len(cheatUsers))
```
